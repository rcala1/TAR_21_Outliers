```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import itertools
from sklearn.preprocessing import StandardScaler
import pickle
```

```python
def load_dataset_numpy(path):
    dataset=pd.read_csv(path)
    return dataset.to_numpy()

def get_Xy_train(train_dataset):
    X=train_dataset[:,1:5]
    y=train_dataset[:,5].astype(int)
    return X,y

def get_train_val_test_split(X,y,train=0.7,val=0.15,test=0.15,random_state=1):
    
    X_train,X_testval,y_train,y_testval=train_test_split(X,y,test_size=val+test,random_state=random_state)
    X_val,X_test,y_val,y_test=train_test_split(X_testval,y_testval,test_size=test/(test+val),random_state=random_state)
    
    return X_train,X_val,X_test,y_train,y_val,y_test
    
def get_acc_f1(y_true,y_pred):
    acc=accuracy_score(y_true,y_pred)
    f1=f1_score(y_true,y_pred)
    return acc,f1

def output_submission(y_submit):
    submission=pd.read_csv('sample_submission.csv')
    submission['is_duplicate']=y_submit
    submission.to_csv('submission.csv',index=False)
    return
```

```python
X,y=get_Xy_train(load_dataset_numpy('train.csv'))
```

```python
X_train,X_val,X_test,y_train,y_val,y_test=get_train_val_test_split(X,y)
```

```python
# Majority classifier
y_pred=[np.argmax(np.bincount(y_test))]*y_test.shape[0]
acc,f1=get_acc_f1(y_test,y_pred)
print(acc,f1)
y_submit=[np.argmax(np.bincount(y_test))]*len(test)
output_submission(y_submit)
```

```python
count_vectorizer = CountVectorizer(max_features=10000) 
```

```python
X_train_string_concat=X_train[:,2:3]+' '+X_train[:,3:4]
X_train_string_concat=X_train_string_concat.squeeze()
```

```python
count_vectorizer.fit(X_train_string_concat)
```

```python
count_vectorizer.get_feature_names()
```

```python
X_train_bag_features_concat=np.concatenate((count_vectorizer.transform(X_train[:,2]).toarray(), \
                                count_vectorizer.transform(X_train[:,3]).toarray()),axis=1)
```

```python
X_val_bag_features_concat=np.concatenate((count_vectorizer.transform(X_val[:,2].astype(str)).toarray(), \
                                count_vectorizer.transform(X_val[:,3].astype(str)).toarray()),axis=1)
```

```python
X_train_and_val_bag_features_concat=np.concatenate((X_train_bag_features_concat,X_val_bag_features_concat))
```

```python
y_train_and_val=np.concatenate((y_train,y_val))
```

```python
X_test_bag_features_concat=np.concatenate((count_vectorizer.transform(X_test[:,2]).toarray(), \
                                count_vectorizer.transform(X_test[:,3].astype(str)).toarray()),axis=1)
```

```python
# BOW with LogisticRegression
scaler=StandardScaler()
logreg=LogisticRegression(verbose=10)
logreg.fit(scaler.fit_transform(X_train_and_val_bag_features_concat),y_train_and_val)
y_pred=logreg.predict(scaler.transform(X_test_bag_features_concat))
acc,f1=get_acc_f1(y_test,y_pred)
print(acc,f1)
```
